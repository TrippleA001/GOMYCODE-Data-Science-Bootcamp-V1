```python
api_key = '85f709c6'

# IMDb IDs of the top 10 movies from the first page
top_10_imdb_ids = ['tt0111161', 'tt0068646', 'tt0071562', 'tt0468569', 'tt0050083',
                   'tt0108052', 'tt0167260', 'tt0110912', 'tt0060196', 'tt0120737']
```

```python
# Initialize a list to store movie details
movie_data = []

# Retrieve detailed information for each IMDb ID
for imdb_id in top_10_imdb_ids:
    movie_url = f'http://www.omdbapi.com/?apikey={api_key}&i={imdb_id}'
    movie_response = requests.get(movie_url)
    movie_details = movie_response.json()

    title = movie_details.get('Title', 'N/A')
    year = movie_details.get('Year', 'N/A')
    director = movie_details.get('Director', 'N/A')
    rating = movie_details.get('imdbRating', 'N/A')

    movie_data.append({'Title': title, 'Year': year, 'Director': director, 'Rating': rating})

# Create a pandas dataframe from the extracted data
df = pd.DataFrame(movie_data)
```
